```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Дані зображення
X = np.array([0, 1, 2, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 2.4, 2.6, 2.8, 3.0])

# Модель лінійної регресії Q = aS + b
model = LinearRegression()
model.fit(X, Y)

# Отримуємо коефіцієнт a (оцінку питомого дебіту)
a = model.coef_
a
```

```python
from sklearn.metrics import mean_squared_error
y_pred = model.predict(X)
mean_squared_error(Y, y_pred)
```

```python
np.sum((y_pred - Y) ** 2)
```
